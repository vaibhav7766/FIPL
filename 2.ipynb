```python
import cv2

path=r"C:\Users\vaibh\OneDrive\Desktop\Wallpapers\1.png"
src=cv2.imread(path)
window_name="Image"

image=cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)

cv2.imshow(window_name,image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
rst,thresh=cv2.threshold(image,127,255,0)
imgRGB=cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
plt.subplot(131),plt.imshow(imgRGB,cmap="gray"),plt.title("Original Image"),plt.axis("off")
plt.subplot(132),plt.imshow(image,cmap="gray"),plt.title("Grayscale Image"),plt.axis("off")
plt.subplot(133),plt.imshow(thresh,cmap="gray"),plt.title("Binary Image"),plt.axis("off")
```
