```python
import cv2
import numpy as np

image=cv2.imread(r"C:\Users\vaibh\OneDrive\Desktop\Wallpapers\wp4082912-jevil-wallpapers.jpg")

height,width=image.shape[:2]

quarter_height, quarter_width=height/4,width/4

T=np.float64([[1,0,quarter_width],[0,1,quarter_height]])

img_translation=cv2.warpAffine(image,T,(width,height))

cv2.imshow("color",image)
cv2.imshow("color1",img_translation)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
center=(width/2,height/2)

rotate_matrix=cv2.getRotationMatrix2D(center=center,angle=45, scale=1)
rotated_image=cv2.warpAffine(src=image, M=rotate_matrix, dsize=(width,height))

print(rotate_matrix)
cv2.imshow("color",image)
cv2.imshow("color1",rotated_image)
cv2.imwrite("image2.jpg",rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

image=cv2.imread(r"C:\Users\vaibh\OneDrive\Desktop\Wallpapers\wp4082912-jevil-wallpapers.jpg")

half=cv2.resize(image,dsize=(0,0),fx=0.5,fy=0.5)
bigger=cv2.resize(image,(1610,1050))
stretch_near=cv2.resize(image,(700,540),interpolation=cv2.INTER_LINEAR)

titles=["Original","Half","stretch_near","bigger"]
images=[image,half,stretch_near,bigger]

for i in range(len(images)):
    plt.subplot(2, 3, i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
    
plt.show()
```
